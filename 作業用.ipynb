```python
fnames = !ls ../../img/patch4/img
len(fnames)
```

```python
import re
import os
import shutil


for fname in fnames:
    pre = re.findall('(.+)_\d+.png', fname)[0]
    suf = re.findall('.+_(\d+.png)', fname)[0]
    if not (os.path.isdir("../../img/patch4/img2/"+pre)): os.mkdir("../../img/patch4/img2/"+pre)
    shutil.copyfile("../../img/patch4/img/"+fname, "../../img/patch4/img2/"+pre+'/'+suf)
```
