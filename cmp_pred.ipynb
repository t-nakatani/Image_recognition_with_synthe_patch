```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
plt.gray()
```

```python
fnames = !ls ../../img/check_prediction
len(fnames)
```

```python
for fname in fnames:
    img1 = cv2.imread('../../img/check_prediction_natural_prob/'+fname)
    cv2.putText(img1, '1', (0, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255, 255, 255), 2, cv2.LINE_AA)
    img2 = cv2.imread('../../img/check_prediction_natural2_prob/'+fname)
    cv2.putText(img2, '2', (0, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255, 255, 255), 2, cv2.LINE_AA)
    img = np.vstack((img1, img2))
    cv2.imwrite('../../img/cmp_pred2/' + fname, img)
```
