```python
pre = [re.findall('(.+)_.+', p)[0] for p in fnames]
suf = [re.findall('.+_(.+)', p)[0] for p in fnames]
```

```python
import collections
c = collections.Counter(pre)
c
```

```python
d2={}
for k,v in c.items():   # 一度pd.Seriesに変換
    d2[k]=pd.Series(v)

df=pd.DataFrame(d2).T
df.head()
```

```python
df[1] = df.index
df_ = pd.DataFrame()
df_['fname'] = df[1]
df_['num_patches'] = df[0]
df_.head()
df_.to_csv('../../img/csv/count_patches.csv', index = False)
```

```python
df__ = df_[df_['num_patches'] > 5]
df__.head()
```

```python
df__['fname'].to_csv('../../img/csv/fnames_more_than_5patches.csv', index = False)
```
