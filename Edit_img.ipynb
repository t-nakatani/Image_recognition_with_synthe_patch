```python
file_names = !ls /Users/nakatani/labo/grabcut/img/mask2

import numpy as np
import cv2
from matplotlib import pyplot as plt
import os

mask = '/Users/nakatani/labo/grabcut/img/mask2/'
resized = '/Users/nakatani/labo/grabcut/img/resized_img2/'
```

```python
def show(img):
    plt.figure(figsize = (20,20))
    plt.imshow(img), plt.show()
    
```

```python
for fname in file_names:
    fname = '42.png'
    print(fname)
    img = cv2.imread(resized+fname)
    mask_img = cv2.imread(mask+fname, cv2.IMREAD_GRAYSCALE)
#     show(img), show(mask_img)
    print(img.shape, mask_img.shape)
    
    _, thresh = cv2.threshold(mask_img, 1, 255, cv2.THRESH_BINARY)

    
    img_ = cv2.bitwise_and(img, img, mask=thresh)
    cv2.imwrite('../../img/foreground2_fixed/'+fname, img_)    
    break
#     show(img_)
#     if fname > '12.png': break
```
